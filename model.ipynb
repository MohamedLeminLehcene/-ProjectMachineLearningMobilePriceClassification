```python
#Import Libraries
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from KNeighbors import KNNClassifier
```

Data Preprocessing

```python
#telecharger dataset
dataset = pd.read_csv("C:/Users/HP/Desktop/MLProject/Impl/Mobile Price Classification/data.csv")
dataset
```

```python
#Vérification des valeurs manquantes
dataset.isnull().sum()
```

```python
#Pour afficher les differents classe
classes = list(set(dataset['price_range'])) 
print('classes:',classes)
```

Splitting data into feature and labels && Lets standardize the data

```python
#pour filter les attributs qui concernant Featuers seulement
X =  dataset.drop(["price_range"],axis=1)
#pour filter l'attribut qui concerne target => y
y = dataset["price_range"]
```

```python
#afficher features => X
X
```

```python
##afficher target => y
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,shuffle=True,random_state=0)
```

```python
X_train = np.asarray(X_train)
X_test = np.asarray(X_test)

y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

```python
#pour desentraliser notre faetures
from sklearn.preprocessing import Normalizer
scaler = Normalizer()

normalized_X_train = scaler.transform(X_train)
normalized_X_test = scaler.transform(X_test)
```

```python
print("X_train before normalize : ")
print(X_train[0:5])

print("X_train after normalize : ")
print(normalized_X_train[0:5])
```

```python
# Calculer la matrice de confusion
def matrixConffudion(y_test_new,y_pred_new):
    conf_matrix = confusion_matrix(y_test_new, y_pred_new)

# Extraire les valeurs de la matrice de confusion
    num_classes = conf_matrix.shape[0]
    tn, fp, fn, tp = [], [], [], []

    if num_classes == 2:
        tn, fp, fn, tp = conf_matrix.ravel()
    else:
        for i in range(num_classes):
            for j in range(num_classes):
                if i == j:
                    tp.append(conf_matrix[i, j])
                else:
                    fn.append(conf_matrix[i, j])
                    fp.append(conf_matrix[j, i])
        tn = sum(conf_matrix.flatten()) - (sum(tp) + sum(fp) + sum(fn))

# Afficher les valeurs
    print('Confusion matrix:\n', conf_matrix)
    print('True Negative (TN):', tn)
    print('False Positive (FP):', fp)
    print('False Negative (FN):', fn)
    print('True Positive (TP):', tp)
```

Load Model => KNN from scratch and sklearn

```python
#KNN 
knn = KNNClassifier(k=4)
#pour entrener notre model from scratch avec normalized_X_train et y_train
knn.fit(normalized_X_train, y_train)
#pour faire la prediction
y_pred_scratch = knn.predict(normalized_X_test)
print(y_pred_scratch)  # Affiche [0, 1]
```

Evaluation Model

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_scratch))
```
